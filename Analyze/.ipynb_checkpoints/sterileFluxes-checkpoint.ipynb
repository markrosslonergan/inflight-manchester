```python
%matplotlib inline
import os, sys, json, pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict as OD
from ipywidgets import interact, interactive, fixed, widgets
from NotebookUtils.ProgressBar import LogProgress as LP
plt.rcParams['font.family']='serif'
plt.rcParams['font.weight']='light'
plt.rcParams['font.size']=14
figsize = (12,8)
```

```python
fluxFiles = ['../FluxFiles/'+name for name in os.listdir('../FluxFiles/') if 'csv' in name[-3:]]
masses = [float('0.'+name.split('_')[-1].split('.')[-2]) for name in fluxFiles]

fig = plt.figure(figsize=figsize)
for fluxFile,mass in zip(fluxFiles,masses):
    flux = pd.read_csv(fluxFile,sep=' ',names=('Energy','Flux'))
    plt.step(flux['Energy'],flux['Flux'],label='%s GeV' %mass)

plt.xlim(0,5)
plt.ylim(1e-16,1e-9)
plt.yscale('log')
plt.grid(ls='--')
plt.legend(loc='best',ncol=2)
plt.ylabel('Flux [a.u.]')
plt.xlabel('Neutrino energy [GeV]')
plt.title('Heavy sterile neutrino fluxes')
plt.show()
```
